```python
%pylab inline
obs_data = loadtxt("movimiento.dat")
x_obs = obs_data[:,0]
y_obs = obs_data[:,1]
scatter(x_obs,y_obs)
```

```python
def likelihood(y_obs, y_model):
    chi = sum((y_obs-y_model)**2)
    return (-chi)
```

```python
def model1(x_obs,a0,a1,a2):
    return a0+a1*x_obs+a2*x_obs**2
```

```python
a0_walk = empty((0))
a1_walk = empty((0))
a2_walk = empty((0))
a0_walk = append(a0_walk, random.random())
a1_walk = append(a1_walk, random.random())
a2_walk = append(a2_walk, random.random())
```

```python
n_iter = 20000

for i in range(n_iter):
    a0_prima = np.random.normal(a0_walk[i],0.1)
    a1_prima = np.random.normal(a1_walk[i],0.1)
    a2_prima = np.random.normal(a2_walk[i],0.1)
    
    y_init = model1(x_obs, a0_walk[i], a1_walk[i], a2_walk[i])
    y_prima = model1(x_obs,a0_prima, a1_prima, a2_prima)
    
    alpha = likelihood(y_obs,y_prima)/likelihood(y_obs,y_init)
    
    if (alpha <= 1.0):
        a0_walk = append(a0_walk,a0_prima)
        a1_walk = append(a1_walk,a1_prima)
        a2_walk = append(a2_walk,a2_prima)
    else:
        beta = random.random()
        if (beta<=exp(-alpha)):
            a0_walk = append(a0_walk,a0_prima)
            a1_walk = append(a1_walk,a1_prima)
            a2_walk = append(a2_walk,a2_prima)
        else:
            a0_walk = append(a0_walk,a0_walk[i])
            a1_walk = append(a1_walk,a1_walk[i])
            a2_walk = append(a2_walk,a2_walk[i])
```
