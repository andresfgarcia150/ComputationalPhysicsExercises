```python
x1_walk = empty((0))
x2_walk = empty((0))
x1_0 = 8.0*((random.random())-0.5)
x2_0 = 8.0*((random.random())-0.5)
x1_walk = append(x1_walk,x1_0)
x2_walk = append(x2_walk,x2_0)
```

```python
n_iter = 20000
sigma = 0.1
for i in range(n_iter):
    x1_prime = np.random.normal(x1_walk[i],sigma)
    x2_prime = np.random.normal(x2_walk[i],sigma)
    alpha = function(x1_prime,x2_prime)/function(x1_walk[i],x2_walk[i])
    if (alpha >= 1.0):
        x1_walk = append(x1_walk,x1_prime)
        x2_walk = append(x2_walk,x2_prime)
    else:
        beta = random.random()
        if (beta <= alpha):
            x1_walk = append(x1_walk,x1_prime)
            x2_walk = append(x2_walk,x2_prime)
        else:
            x1_walk = append(x1_walk,x1_walk[i])
            x2_walk = append(x2_walk,x2_walk[i])
```
